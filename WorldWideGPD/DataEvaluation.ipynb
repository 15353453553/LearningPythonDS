```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datasheet = pd.merge( pd.read_csv('WorldWideHomeless'), (pd.read_csv('WorldWideGPD')), on='Countries')
datasheet.pop('Unnamed: 0_x')
datasheet.pop('Unnamed: 0_y')
```

```python
datasheet
```

```python
x = datasheet['Homeless_per_10k']
y = datasheet['GPD_per_capita']
c = x
xlim = 2000
ylim = 100000

af = plt.figure()

axes = af.add_axes([0, 0, 2, 4])

axes.scatter(x,y, marker='o', c=c, s=20)

y2 = np.linspace(0,100000,100)
x2 = np.ones(100)*60

axes.plot(x2,y2,'r--')
axes.set_xlabel('Homeless per 10k')
axes.set_ylabel('GPD per capita')
axes.set_xlim(right=xlim)
axes.set_ylim(top=ylim)

for i in range(datasheet.shape[0]):
    if(x[i]<xlim and y[i]<ylim):
        axes.text(x=x[i]+0.3,y=y[i]+0.3,s=datasheet.Countries[i], 
                  fontdict=dict(color="black",size=9))
```
